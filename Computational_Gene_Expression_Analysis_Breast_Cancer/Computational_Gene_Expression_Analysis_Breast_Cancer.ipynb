```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)

# Simulated gene expression data
genes = [f"Gene{i}" for i in range(100)]
data = pd.DataFrame({
    "Normal": np.random.normal(5, 1, 100),
    "Cancer": np.random.normal(6, 1.2, 100)
}, index=genes)

data["logFC"] = data["Cancer"] - data["Normal"]
data["pval"] = np.random.uniform(0, 0.05, 100)
data["neglogp"] = -np.log10(data["pval"])
```

```python
plt.scatter(data["logFC"], data["neglogp"])
plt.axvline(x=1, color='r')
plt.axvline(x=-1, color='r')
plt.title("Volcano Plot (Gene Expression)")
plt.xlabel("Log Fold Change")
plt.ylabel("-log10(p-value)")
plt.show()
```

```python
plt.figure(figsize=(6,5))
plt.scatter(data["logFC"], data["neglogp"], alpha=0.6)

plt.axvline(1, color='red', linestyle='--')
plt.axvline(-1, color='red', linestyle='--')

plt.title("Differential Gene Expression (Volcano Plot)")
plt.xlabel("Log Fold Change")
plt.ylabel("-log10(p-value)")
plt.show()
```

```python
sns.heatmap(data[["Normal","Cancer"]].head(20), cmap="coolwarm")
plt.title("Gene Expression Heatmap")
plt.show()
```

```python
data_sorted = data.sort_values("logFC", ascending=False)
top_genes = data_sorted.head(10)
top_genes
```

```python
# Select top genes
top_genes = data.sort_values("logFC", ascending=False).head(10)
top_genes
```
