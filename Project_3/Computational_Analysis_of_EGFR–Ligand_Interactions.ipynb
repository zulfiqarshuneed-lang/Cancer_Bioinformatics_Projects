```python
import pandas as pd

data = pd.DataFrame({
    "Ligand": ["Curcumin", "Quercetin", "Resveratrol"],
    "Binding Energy (kcal/mol)": [-8.4, -9.1, -7.8]
})

data
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.bar(data["Ligand"], data["Binding Energy (kcal/mol)"])
plt.title("Ligand Binding Affinity to EGFR")
plt.xlabel("Ligand")
plt.ylabel("Binding Energy (kcal/mol)")
plt.show()
```

```python
data.to_csv("docking_results.csv", index=False)
```
